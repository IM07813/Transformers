```python
from TTS.api import TTS
from pydub import AudioSegment

# Initialize TTS with XTTS v2 model
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2", gpu=True)

# Split the text into two chunks at a natural break point
chunk1 = """Your text here less than 300 words in every chunck"""
chunk2 = """Your text here less than 300 words in every chunck"""

# Process each chunk separately
tts.tts_to_file(
    text=chunk1,
    file_path="temp_part1.wav",
    speaker_wav="your source file that you want to clone .wav file extention",
    language="en"
)

tts.tts_to_file(
    text=chunk2,
    file_path="temp_part2.wav",
    speaker_wav="your source file that you want to clone .wav file extention",
    language="en"
)

# Combine the audio files seamlessly
def combine_wav_files(files, output_file):
    combined = AudioSegment.empty()
    for file in files:
        audio = AudioSegment.from_wav(file)
        combined += audio
    
    # Export with the same parameters as the input files to maintain quality
    combined.export(output_file, format="wav")

# Combine the files
combine_wav_files(["temp_part1.wav", "temp_part2.wav"], "output_speech_v1.wav")

# Clean up temporary files
import os
os.remove("temp_part1.wav")
os.remove("temp_part2.wav")
```
